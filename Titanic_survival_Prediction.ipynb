```python

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
a=pd.read_csv('/content/Titanic-Dataset.csv')
print(a)
```

```python
a.head()
```

```python
a.info()
```

```python
a = a.drop(['PassengerId' , 'Name' ,'Cabin'] , axis = 1)
a.head(5)
```

```python
a['Age'] = a['Age'].fillna(a['Age'].mean())
```

```python
a.dropna(inplace=True)
```

```python
B = LabelEncoder()
a['Sex'] = B.fit_transform(a['Sex'])
a['Embarked'] = B.fit_transform(a['Embarked'])
```

```python
a = a.drop(['Ticket'] , axis = 1)
```

```python
a['Fare'] = (a['Fare'] - a['Fare'].min()) / (a['Fare'].max() - a['Fare'].min())
```

```python
a.head(10)
```

```python
a['Survived'].value_counts()
```

```python
X = a.drop('Survived' ,axis = 1)
y = a['Survived']
```

```python
X_train ,X_test ,y_train ,y_test = train_test_split(X ,y ,test_size = 0.2 ,random_state = 0)
```

```python
b = RandomForestClassifier(n_estimators=100, random_state=42)
b.fit(X_train, y_train)
```

```python
y_pred = b.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(' ')
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', conf_matrix)
print(' ')
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)
```

Logistic Regression

```python
c = LogisticRegression(max_iter=1000, random_state=42)
c.fit(X_train, y_train)
```

```python
y_pred = c.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(' ')
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', conf_matrix)
print(' ')
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)
```
